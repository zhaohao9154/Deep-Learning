```python
import os
import pandas as pd
import urllib.request
```

```python
date_range = pd.date_range(start='09/01/2016', end='09/01/2017').strftime("%Y%m%d")
stem = list(date_range)
time = ["%03d" %i for i in list(range(8,24))] 
missing_dates = []
```

```python
for date in stem:
    html = 'https://www2.mmm.ucar.edu/projects/ncar_ensemble/archive/'+ date
    for hr in time:
        new_html = html+'00/NE/t2_mean_f'+ hr +'_NE.png'
        file_name = 'tmpf_' + date +'00_f'+ hr + '.jpg'
        try:
            urllib.request.urlretrieve(new_html, file_name)
        except:
            missing_dates.append('Date:' + date +'Time:'+ hr)
```

```python
missing_dates
```

```python
with open("missing_dates.txt", "w") as file:
    file.write(str(missing_dates))
```
